```python
import torch
```

```python
empty_state_dict = torch.load(empty_model_pth,map_location="cpu")['state_dict']
trained_state_dict = torch.load(trained_model_pth,map_location="cpu")['state_dict']
```

```python
for key in trained_state_dict.keys():
    new_key = "backbone." + key[7:]
    if new_key not in empty_state_dict.keys():
        print(new_key)
    else:
        if trained_state_dict[key].shape != empty_state_dict[new_key].shape:
            print(key, new_key)

for new_key in empty_state_dict.keys():
    # key = new_key.replace("backbone.", "module.")
    key = "module." + new_key[9:]
    if key not in trained_state_dict.keys():
        print(new_key, empty_state_dict[new_key].shape)

# this is very lightweight
```

```python
new_save = {"state_dict": {}}
for key in trained_state_dict.keys():
    if not key[:7] == "module.":
        print(key)
    else:
        new_key = "backbone." + key[7:]
        if new_key not in empty_state_dict.keys():
            print(new_key)
        elif trained_state_dict[key].shape != empty_state_dict[new_key].shape:
            print(key, new_key)
        else:
            new_save['state_dict'][new_key] = trained_state_dict[key]
```
